#Initiation

```r
#Building My Own:
n<- 500 #This is the number of steps.
#Defining the movement in the x and y direction:
stepsx<- sample(c(-1,1),n,replace=TRUE)
stepsy<- sample(c(-1,1),n,replace=TRUE)

#now we want the running total of the values for the random walk to calculate
#the positions:

x<-cumsum(stepsx)
y<-cumsum(stepsy)

#Finally the plot:
plot(x,y,col="blue", main="Random Walk with 500 steps", type="l")
```
